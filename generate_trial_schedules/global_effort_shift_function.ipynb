```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Parameters
what_is_displayed = np.arange(0, 101, 1)  # Updated to include all values from 0 to 100
average_reward = 60
net_value_shift = 30
k = 1.1

# Net value function
what_is_displayed_level = what_is_displayed / 10
discounted_value = average_reward - k * what_is_displayed_level**2

# Shifted effort 
what_is_displayed_level = what_is_displayed / 10
what_participant_exerts_level = np.sqrt((k * what_is_displayed_level**2 + net_value_shift) / k)
what_participant_exerts = what_participant_exerts_level * 10

# Plotting what_participant_exerts vs what_is_displayed (and a straight line for comparison to show that effort gets shifted less for higher effort levels)
plt.figure()
plt.plot(what_is_displayed, what_participant_exerts)
plt.ylim(50, 115)
plt.xlim(50, 100)
plt.xlabel('Displayed effort', fontsize=24) 
plt.ylabel('Exerted effort', fontsize=24)   
plt.title('Shifted', fontsize=26) 
plt.grid(True)
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x)}%'))
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: f'{int(y)}%'))

# Increase the size of tick labels
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

```python
import numpy as np
import pandas as pd

# Parameters
what_participant_exerts = np.arange(0, 101, 1)
net_value_shift = 30
k = 1.1

# Calculating what_is_displayed from what_participant_exerts
what_participant_exerts_level = what_participant_exerts / 10
what_is_displayed_level = np.sqrt((k * what_participant_exerts_level**2 - net_value_shift) / k)
what_is_displayed = what_is_displayed_level * 10
what_is_displayed = np.maximum(what_is_displayed, 0)

# Creating a DataFrame for the corresponding values
df_corresponding_values = pd.DataFrame({
    'What Participant Exerts (%)': what_participant_exerts,
    'What Is Displayed (%)': what_is_displayed
})

# Display the DataFrame
df_corresponding_values[::10]
```
