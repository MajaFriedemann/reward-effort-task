```python
import numpy as np
import pandas as pd

# Parameters
what_participant_exerts = np.arange(0, 101, 1)
net_value_shift = 30
k = 1.1

# Calculating what_is_displayed from what_participant_exerts
what_participant_exerts_level = what_participant_exerts / 10
what_is_displayed_level = np.sqrt((k * what_participant_exerts_level**2 - net_value_shift) / k)
what_is_displayed = what_is_displayed_level * 10
what_is_displayed = np.maximum(what_is_displayed, 0)

# Creating a DataFrame for the corresponding values
df_corresponding_values = pd.DataFrame({
    'What Participant Exerts (%)': what_participant_exerts,
    'What Is Displayed (%)': what_is_displayed
})

# Display the DataFrame
df_corresponding_values[::10]
```
